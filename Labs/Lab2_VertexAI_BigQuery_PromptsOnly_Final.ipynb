## Part D — Common Table Expressions (CTEs)
**Aim:** Make complex logic readable and testable in steps.

### D1. Multi‑step ranking with CTEs
**Prompt:**
```
BigQuery SQL only.
Goal: Within each `Region`, rank states by total sales and return top 3 per region.
Table: `mgmt-467-project-1.superstore_data.sales`
CTE 1 (`state_sales`): SUM(Sales) by `Region`, `State`
CTE 2 (`ranked_state_sales`): Add `RANK() OVER (PARTITION BY Region ORDER BY total_sales DESC)` as `sales_rank`
Final SELECT: rows where `sales_rank <= 3`
Output columns: `Region`, `State`, `total_sales`, `sales_rank`
Sort: by `Region`, then `sales_rank`
```
**Ask for**: a one-paragraph explanation of each step, then **provide only the final runnable SQL**.

Here are the explanations for each step of the query:

The first CTE, `state_sales`, calculates the total sales for each combination of `region` and `state` by grouping the data from the `superstore_clean` view. This step provides a summarized view of sales performance at the state level within each region.

The second CTE, `ranked_state_sales`, takes the results of `state_sales` and applies a ranking within each `region`. The `RANK() OVER (PARTITION BY region ORDER BY total_sales DESC)` window function assigns a rank to each state within its respective region based on `total_sales` in descending order, meaning the state with the highest total sales in a region gets rank 1.

Finally, the main `SELECT` statement retrieves the `region`, `state`, `total_sales`, and `sales_rank` from the `ranked_state_sales` CTE, filtering the results to include only those rows where the `sales_rank` is less than or equal to 3. This effectively gives you the top 3 states by total sales within each region, and the results are ordered by region and then by their sales rank.

````python
query_string = """
WITH state_sales AS (
    SELECT
        region,
        state,
        SUM(sales) AS total_sales
    FROM
        `mgmt-467-project-1.lab1_foundations.superstore_clean`
    GROUP BY
        region,
        state
),
ranked_state_sales AS (
    SELECT
        region,
        state,
        total_sales,
        RANK() OVER (PARTITION BY region ORDER BY total_sales DESC) AS sales_rank
    FROM
        state_sales
)
SELECT
    region,
    state,
    total_sales,
    sales_rank
FROM
    ranked_state_sales
WHERE
    sales_rank <= 3
ORDER BY
    region,
    sales_rank;
"""

# Execute the query
query_job = client.query(query_string)
results_df = query_job.to_dataframe()

# Display the results
display(results_df)
````

## Part E — Window Functions (ROW_NUMBER, RANK, DENSE_RANK, LAG/LEAD, moving averages)
**Aim:** Compare rows across partitions and time; compute trends and ranks without collapsing rows.

### E1. Top product per region (ROW_NUMBER)
**Prompt:**
```
BigQuery SQL only.
Task: For each `Region`, return only the single highest-revenue `Sub_Category`.
Table: `mgmt-467-project-1.superstore_data.sales`
CTE `subcat_sales`: SUM(Sales) by `Region`, `Sub_Category`
Add `ROW_NUMBER() OVER (PARTITION BY Region ORDER BY total_sales DESC)` as rn
Final: filter `rn = 1`
Output: `Region`, `Sub_Category`, `total_sales`
Sort by `Region`
```
**Why `ROW_NUMBER` instead of `RANK`?** Ask the model to add a 2-sentence contrast.

`ROW_NUMBER()` assigns a unique sequential integer rank to each row within its partition, even if there are ties in the ordering column. `RANK()`, on the other hand, assigns the same rank to rows with the same value in the ordering column, and then skips the subsequent rank(s).

````python
query_string = """
WITH subcat_sales AS (
    SELECT
        region,
        sub_category,
        SUM(sales) AS total_sales
    FROM
        `mgmt-467-project-1.lab1_foundations.superstore_clean`
    GROUP BY
        region,
        sub_category
),
ranked_subcat_sales AS (
    SELECT
        region,
        sub_category,
        total_sales,
        ROW_NUMBER() OVER (PARTITION BY region ORDER BY total_sales DESC) AS rn
    FROM
        subcat_sales
)
SELECT
    region,
    sub_category,
    total_sales
FROM
    ranked_subcat_sales
WHERE
    rn = 1
ORDER BY
    region;
"""

# Execute the query
query_job = client.query(query_string)
results_df = query_job.to_dataframe()

# Display the results
display(results_df)
````

### E2. YoY growth with LAG
**Prompt:**
```
BigQuery SQL only.
Task: Compute year-over-year revenue growth for 'Phones' sub-category.
Table: `mgmt-467-project-1.superstore_data.sales`
Steps:
- Filter to `Sub_Category = 'Phones'`
- Aggregate yearly revenue using EXTRACT(YEAR FROM Order_Date)
- Add `LAG(yearly_revenue) OVER (ORDER BY year)` as `prev_revenue`
- Compute `yoy_pct = 100.0 * (yearly_revenue - prev_revenue) / prev_revenue`
Output: `year`, `yearly_revenue`, `prev_revenue`, `yoy_pct`
Sort by `year` ASC
```
**Ask for**: a guard against divide-by-zero or NULL previous year.

````python
query_string = """
WITH yearly_phones_sales AS (
    SELECT
        EXTRACT(YEAR FROM order_date) AS order_year,
        SUM(sales) AS yearly_revenue
    FROM
        `mgmt-467-project-1.lab1_foundations.superstore_clean`
    WHERE
        sub_category = 'Phones'
    GROUP BY
        order_year
),
lagged_sales AS (
    SELECT
        order_year,
        yearly_revenue,
        LAG(yearly_revenue) OVER (ORDER BY order_year) AS prev_revenue
    FROM
        yearly_phones_sales
)
SELECT
    order_year AS year,
    yearly_revenue,
    prev_revenue,
    SAFE_DIVIDE( (yearly_revenue - prev_revenue) * 100.0, prev_revenue) AS yoy_pct
FROM
    lagged_sales
ORDER BY
    year ASC;
"""

# Execute the query
query_job = client.query(query_string)
results_df = query_job.to_dataframe()

# Display the results
display(results_df)
````

### F2. Reduce cost / improve speed
**Prompt:**
```
Act as a BigQuery cost optimizer.
Given this query (below), list 3 ways to reduce scanned bytes and improve performance without changing the business logic.
[SELECT
    AVG(monthly_revenue) AS average_monthly_revenue
FROM (
    SELECT
        DATE_TRUNC(order_date, MONTH) AS month,
        SUM(sales) AS monthly_revenue
    FROM
        `mgmt-467-project-1.lab1_foundations.superstore_clean`
    GROUP BY
        DATE_TRUNC(order_date, MONTH)
);]
Prioritize: partition filters, column pruning, pre-aggregations, and temporary results via CTEs.
```

Here are three ways to reduce scanned bytes and potentially improve the performance of the provided query without changing its business logic, keeping in mind your prioritization of partition filters, column pruning, pre-aggregations, and CTEs:

1.  **Partitioning the Source Table:** If the underlying `superstore` table (from which `superstore_clean` is derived) were partitioned by `Order_Date` (or a date/timestamp column), you could filter on the partition column in your query's `WHERE` clause. While this specific query calculates the average over the *entire* dataset's date range, if you ever needed to calculate the average monthly revenue for a *specific* time period (e.g., a particular year), partitioning would allow BigQuery to only scan the relevant partitions, significantly reducing scanned bytes.

2.  **Column Pruning (Already Applied):** The current query already does a good job of column pruning. It only selects `order_date` and `sales` from the `superstore_clean` view in the inner query. Since the `superstore_clean` view itself selects all columns from the original `superstore` table, true column pruning would ideally happen at the source table level if you were querying it directly or if the view were more restrictive. However, within the context of this query against the view, you are already only selecting the necessary columns (`order_date` and `sales`) for the calculation, avoiding scanning unnecessary columns like `customer_name` or `product_name`.

3.  **Materialized View (Pre-aggregation):** For a query like this that calculates an aggregate (`SUM(sales)`) grouped by a dimension (`DATE_TRUNC(order_date, MONTH)`), creating a materialized view could significantly improve performance and reduce cost on subsequent runs.
